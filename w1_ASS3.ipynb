```python
def solve_fractional_knapsack():
    T = int(input())
    results = []

    for _ in range(T):
        N, W = map(int, input().split())
        items = []
        for _ in range(N):
            v, w = map(int, input().split())
            items.append((v, w))

        # Calculate value-to-weight ratio and store as (ratio, value, weight)
        # Sorting in descending order of ratio
        items.sort(key=lambda x: x[0] / x[1], reverse=True)

        current_weight = 0
        current_value = 0.0

        for v, w in items:
            if current_weight + w <= W:
                # Take the whole item
                current_weight += w
                current_value += v
            else:
                # Take a fraction of the item
                remaining_capacity = W - current_weight
                fraction = remaining_capacity / w
                current_value += fraction * v
                current_weight = W  # Knapsack is now full
                break
        results.append(current_value)

    for res in results:
        print(f"{res:.6f}")

solve_fractional_knapsack()
```
