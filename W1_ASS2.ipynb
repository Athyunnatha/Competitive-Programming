```python
def find_first_occurrence(arr, target):
    low, high = 0, len(arr) - 1
    first_occurrence = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            first_occurrence = mid
            high = mid - 1  # Try to find an even earlier occurrence
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return first_occurrence

def find_last_occurrence(arr, target):
    low, high = 0, len(arr) - 1
    last_occurrence = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            last_occurrence = mid
            low = mid + 1  # Try to find an even later occurrence
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return last_occurrence

def solve():
    N, Q = map(int, input().split())
    product_ids = list(map(int, input().split()))

    results = []
    for _ in range(Q):
        query_id = int(input())
        first = find_first_occurrence(product_ids, query_id)

        if first == -1:
            results.append(0)
        else:
            last = find_last_occurrence(product_ids, query_id)
            results.append(last - first + 1)

    return results

T = int(input())
all_test_case_results = []
for _ in range(T):
    all_test_case_results.extend(solve())

for res in all_test_case_results:
    print(res)
```
