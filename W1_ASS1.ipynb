```python
def solve():
    N = int(input())
    meetings = []
    for _ in range(N):
        s, e = map(int, input().split())
        meetings.append((s, e))

    meetings.sort(key=lambda x: x[1])

    count = 0
    last_end_time = -1

    for start, end in meetings:

        if start > last_end_time:
            count += 1
            last_end_time = end
    print(count)

T = int(input())

for _ in range(T):
    solve()
```
